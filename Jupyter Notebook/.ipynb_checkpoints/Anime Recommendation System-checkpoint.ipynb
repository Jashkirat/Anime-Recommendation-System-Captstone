### Correlation Analysis

```python
# Only the numerical columns
numerical_df = Merged_Data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot heatmap
fig = plt.figure(figsize=(15, 15))
ax = fig.gca()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='PRGn')
plt.savefig('correlation_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

```

Green squares indicate a positive correlation between variables, suggesting that as one variable increases, the other tends to increase as well. The intensity of the green color signifies the strength of this positive correlation.

Purple squares denote a negative correlation, implying that as one variable increases, the other tends to decrease. Darker shades of purple indicate a stronger negative correlation.

White squares typically represent no correlation, indicating little to no linear relationship between the variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define categories and colors
categories = ['TV', 'Movie', 'OVA', 'Special', 'ONA', 'Music']
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'purple', 'yellow']

# Loop through each category and create separate plots
for i, category in enumerate(categories):
    print(f"Plotting distribution for {category} category...")
    category_data = Merged_Data[Merged_Data['category'] == category]
    
    # Set up the subplot layout
    fig, axs = plt.subplots(1, 2, figsize=(20, 8), sharex=False, sharey=False)
    plt.tight_layout(pad=4.0)
    
    # Plotting histogram for average ratings
    sns.histplot(category_data['rating'], color=colors[i], kde=True, bins=20, alpha=0.5, ax=axs[0])
    axs[0].set_title(f"\nAnime's Average Ratings Distribution [{category}]\n", fontsize=20)
    axs[0].set_xlabel("Rating")
    axs[0].set_ylabel("Total")
    
    # Plotting histogram for user ratings
    sns.histplot(category_data['user_rating'], color=colors[i], kde=True, bins=20, alpha=0.5, ax=axs[1])
    axs[1].set_title(f"\nUsers Anime Ratings Distribution [{category}]\n", fontsize=20)
    axs[1].set_xlabel("Rating")
    axs[1].set_ylabel("Total")
    
    # Save the plot as an image
    plt.savefig(f'Distribution_of_{category}s.png', dpi=300, bbox_inches='tight')
    
    # Show the plot
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define genres and colors
genres = ['TV', 'Movie', 'OVA', 'Special', 'ONA', 'Music']
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'purple', 'yellow']

print("Let's explore the ratings distribution of TV category :\n")

# Set up the subplot layout
fig, axs = plt.subplots(1, 2, figsize=(20, 8), sharex=False, sharey=False)
plt.tight_layout(pad=4.0)

# Plotting histogram for average ratings with legend
for i, genre in enumerate(genres):
    genre_data = Merged_Data[Merged_Data['category'] == genre]
    sns.histplot(genre_data['rating'], color=colors[i], kde=True, bins=20, alpha=0.5, ax=axs[0], label=genre)
axs[0].set_title("\nAnime's Average Ratings Distribution All Category\n", fontsize=20)
axs[0].set_xlabel("Rating")
axs[0].set_ylabel("Total")
axs[0].legend()

# Plotting histogram for user ratings with legend
for i, genre in enumerate(genres):
    genre_data = Merged_Data[Merged_Data['category'] == genre]
    sns.histplot(genre_data['user_rating'], color=colors[i], kde=True, bins=20, alpha=0.5, ax=axs[1], label=genre)
axs[1].set_title("\nUsers Anime Ratings Distribution Category \n", fontsize=20)
axs[1].set_xlabel("Rating")
axs[1].set_ylabel("Total")
axs[1].legend()
plt.savefig('Distribution_of_All_Category.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Extracting the genre column from the DataFrame
Anime_Temp = Merged_Data[['genre']]

# Splitting the genre values by commas and spaces
Anime_Temp.loc[:, 'genre'] = Anime_Temp['genre'].str.split(", | , | ,")

# Expanding the DataFrame to have one row for each genre
Anime_Temp = Anime_Temp.explode('genre')

# Capitalizing the genre names
Anime_Temp['genre'] = Anime_Temp['genre'].str.title()

# Creating a DataFrame to count the occurrences of each genre
genre_counts = Anime_Temp['genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Occurrences']

# Printing the top 5 genres and their occurrences
print('Top 5 genres and their occurrences:')
print(genre_counts.head(5))
```

```python
# Selecting top 10 genres
top_10_genres = genre_counts.head(10)

print("Let's explore the Top 10 Anime Generes :\n")

# Plotting a hollow pie chart with style
plt.figure(figsize=(10, 6))
plt.pie(top_10_genres['Occurrences'], labels=top_10_genres['Genre'], autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors, wedgeprops=dict(width=0.3, edgecolor='white'))
plt.title('Top 10 Anime Genres', fontsize=16, fontweight='bold', color='navy')
plt.axis('equal')
plt.savefig('Top_10_Anime_Generes.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Data Cleaning

```python
selected_users = Merged_Data['user_id'].value_counts()
Merged_Data = Merged_Data[Merged_Data['user_id'].isin(selected_users[selected_users >= 25].index)]
```

#### Only Keeping Users Who Have Rated more than 25 Anime so that we can remove bias reviews

```python
Merged_Data.head()
```

```python
import pandas as pd
import string
import re

def clean_text(text):
    # Remove punctuation using regex
    text = re.sub(r'[^\w\s]', '', text)
    
    # Convert text to lowercase
    text = text.lower()
    
    # Remove non-alphanumeric characters and spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    return text


Merged_Data.loc[:, 'name'] = Merged_Data['name'].apply(clean_text)
```
